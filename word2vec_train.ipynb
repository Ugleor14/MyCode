добавим библиотеку для обработки текста и создадим метод *чистящий* слова в предложениях
*в идеале заменить на nltk библиотеку : https://www.kaggle.com/alxmamaev/how-to-easy-preprocess-russian-text 
How to easy preprocess Russian text*

```python
import re #библиотека для работы с текстами
def ClearText(text):
    text = text.lower().replace("ё", "е")
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', '', text)
    text = re.sub('@[^\s]+', ' ', text)
    text = re.sub('[^a-zA-Zа-яА-Я1-9]+', ' ', text)
    text = re.sub(' +', ' ', text)
    return text.strip()
```

### Алгоритмы Skip-Gram or CBOW

- **Skip-gram** : использование центрального слова для предсказывания окружающих
- **CBOW** : использование окружающих слов для предсказывания центрального

### Скользящее окно : Window
- по предложению движется окнов состоящие из n cлов , поочередно добавляющее следующее и убирающее предыдущее
в этом окне анализируется вероятность встретить слово по его окружению с ипользованием одного из алгоритмов

```python
#блок тренировки модели
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO) #функция логгирования
# Считываем файл с предобработанными твитами
data = gensim.models.word2vec.LineSentence('data/tweets.txt')
# Обучаем модель 
model = Word2Vec(
    data,                  #файл обучения
    size         = 200,       #размерность признакового пространства, размер вектора слова (рекоменд 300) ;
    window       = 5,         #количество слов в скользещем окне (рекоменд 1/5/8/10);
    #sg           = 0,         #алгоритм обучения 1 = scip-gram , 0 = CBOW ;
    #hs           = 0,         #использование иархического softmax ;
    min_count    = 3,         #минимальная частота слов в словаре ;
    #negative     = 5,         #негативное семлирование ;
    #iter         = 5,         #количество эпох (рекоменд 3, подбираеся в ручную);
    #sorted_vocab = 0,         #слова в словаре индексируются в порядке убывания частот (1:вкл 0:выкл)
    #alpha        = 0.025,      #lerning rate (рек от 0.025 до 0.5)
    workers      = multiprocessing.cpu_count(),#функция для использования больших мощностей процессора 
)
model.save("model/model.w2v") 
```
